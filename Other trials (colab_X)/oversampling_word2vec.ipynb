```python
import imblearn
import pandas as pd
from pandas import read_csv
from imblearn.over_sampling import SMOTE
from collections import Counter
from matplotlib import pyplot
from sklearn.preprocessing import LabelEncoder
```

```python
# https://machinelearningmastery.com/multi-class-imbalanced-classification/
# load the csv file as a data frame
train_df = pd.read_csv("wordEmbeddingTrain.csv")
del train_df['index']
print('Shape of the train dataset: ' + str(train_df.shape))
train_df.head()
```

```python
# split into input and output elements
X= train_df.iloc[:,0:100].values 
y= train_df["labels"]
```

```python
# label encode the target variable
enc = LabelEncoder().fit(y)
y = enc.transform(y)
```

```python
# summarize distribution
counter = Counter(y)
for k,v in counter.items():
    per = v / len(y) * 100
    print('Class=%d, n=%d (%.3f%%)' % (k, v, per))
```

```python
# plot the distribution
pyplot.bar(counter.keys(), counter.values())
pyplot.show()
```

```python
# transform the dataset
oversample = SMOTE()
X, y = oversample.fit_resample(X, y)
```

```python
# summarize distribution
counter = Counter(y)
for k,v in counter.items():
    per = v / len(y) * 100
    print('Class=%d, n=%d (%.3f%%)' % (k, v, per))
```

```python
# plot the distribution
pyplot.bar(counter.keys(), counter.values())
pyplot.show()
```

```python
train_df_new = pd.DataFrame(X)
y_label = enc.inverse_transform(y)
train_df_new['labels'] = y_label
```
